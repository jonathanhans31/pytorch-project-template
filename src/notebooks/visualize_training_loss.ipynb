```python
import numpy as np
import matplotlib.pyplot as plt
from glob import glob

import sys
import yaml
PATHS = yaml.safe_load(open("../../paths.yaml"))
for k in PATHS: sys.path.append(PATHS[k])
```

```python
paths = glob("../../experiments/416936859841431826/*/metrics/l2_loss")
```

```python
losses = []
for p in paths:
    with open(p) as f:
        loss = np.array([float(x.split(" ")[1].strip()) for x in f.readlines()])
        losses.append(loss)
losses = np.stack(losses)
```

```python
plt.rcParams["figure.figsize"] = (4,3)

plt.plot(losses.T)
plt.legend(["base.yaml", "base2.yaml"])
plt.title("Losses of base.yaml vs base2.yaml")
plt.show()
```
